```python
!pip install tensorflow

import tensorflow as tf

# 1. Create the Model (A single Dense layer)
# Input shape is [1] because we are passing one number at a time
model = tf.keras.Sequential([
    tf.keras.Input(shape=[1]), # Recommended way to specify input shape
    tf.keras.layers.Dense(units=1)
])

# 2. Compile: Pick an optimizer (the math teacher) and loss (the grading scale)
model.compile(optimizer='sgd', loss='mean_squared_error')

# 3. Data: y = 2x - 1
xs = tf.constant([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = tf.constant([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

# 4. Train: This is where backpropagation happens automatically
history = model.fit(xs, ys, epochs=500, verbose=1)

print(model.predict(tf.constant([[10.0]], dtype=float))) # Should be close to 19.0
```

```python
print(f"Learned Weight (W): {model.get_weights()[0]}")
print(f"Learned Bias (b): {model.get_weights()[1]}")
```

```python
print(f"Learned Weight (W): {model.get_weights()}")
```

```python
from numpy import log10
import matplotlib.pyplot as plt
plt.plot(log10(history.history['loss']))
plt.xlabel('Epoch')
plt.ylabel('log of Loss')
plt.title('Backpropagation in Action')
plt.show()
```
